```python
from datasets import load_metric

bleu = load_metric("bleu")

# predictions and references need to be lists of lists
predictions = [["This is a prediction."], ["Another prediction."]]
references = [[["This is a reference."]], [["Another reference."]]]

result = bleu.compute(predictions=predictions, references=references)
print(result)
```

```python
import os
import pandas as pd
import pyarrow.parquet as pq

# Parquet 파일들이 저장된 디렉토리 경로
parquet_dir = r'C:\Users\wjdrb\vscode_code\MathBridge_new\data\gpt_parquet_files'

# 디렉토리 안의 모든 parquet 파일 목록 가져오기
parquet_files = [os.path.join(parquet_dir, f) for f in os.listdir(parquet_dir) if f.endswith('.parquet')]
dataframes = [pq.read_table(file).to_pandas() for file in parquet_files]
combined_dataframe = pd.concat(dataframes, ignore_index=True)
combined_dataframe.to_parquet('gpt_generated.parquet')

print("All Parquet files have been combined into 'combined_parquet_file.parquet'")
```

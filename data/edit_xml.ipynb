```python
import xml.etree.ElementTree as ET
import pandas as pd

def parse_xml_to_dataframe(xml_file):
    # XML 파일 파싱
    tree = ET.parse(xml_file)
    root = tree.getroot()
    
    # XML 데이터를 저장할 리스트
    data = []
    
    # XML 파일의 각 요소를 순회하며 데이터 추출
    for child in root:
        row_data = {}
        for element in child:
            row_data[element.tag] = element.text
        data.append(row_data)
    
    # 리스트를 DataFrame으로 변환
    df = pd.DataFrame(data)
    return df

# XML 파일 경로
xml_file_path = r'C:\Users\wjdrb\vscode_code\MathBridge_new\data\downloads\arXiv_src_manifest.xml'

# DataFrame으로 변환
df = parse_xml_to_dataframe(xml_file_path)
```

```python
import pandas as pd
import xml.etree.ElementTree as ET

def dataframe_to_xml(df, root_tag, row_tag, xml_file_path):
    # XML 루트 요소 생성
    root = ET.Element(root_tag)
    
    # DataFrame의 각 행을 XML 요소로 변환
    for _, row in df.iterrows():
        row_element = ET.SubElement(root, row_tag)
        for col in df.columns:
            cell = ET.SubElement(row_element, col)
            cell.text = str(row[col])
    
    # 트리 구조 생성
    tree = ET.ElementTree(root)
    
    # XML 파일로 저장
    tree.write(xml_file_path, encoding='utf-8', xml_declaration=True)




# DataFrame을 XML 파일로 변환
xml_file_path = 'arXiv_src_manifest_edited.xml'
dataframe_to_xml(df, 'People', 'Person', xml_file_path)

print(f"DataFrame을 {xml_file_path}로 변환했습니다.")
```

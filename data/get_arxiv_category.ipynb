```python
import requests
from bs4 import BeautifulSoup

# URL of the page you want to scrape
url = 'https://arxiv.org/category_taxonomy'

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content of the page using Beautiful Soup
soup = BeautifulSoup(response.text, 'html.parser')

# Navigate through the tags: main > div > ul > li > b
# Assuming the structure is consistent and these tags directly follow each other
categories = [h4.text.strip() for h4 in soup.find_all('h4')]


```

```python
import re

def remove_text_within_parentheses(input_list):
    # Pattern to find text within parentheses
    pattern = re.compile(r'\(.*?\)')
    
    # Replace the matched text with an empty string
    return [pattern.sub('', item).strip() for item in input_list]
```

```python
categories = remove_text_within_parentheses(categories)
```

```python
left_strings = [item.split('.')[0] for item in categories]
category = list(set(left_strings))
category
```

```python
input_list = categories
input_list
```

```python
# 주어진 리스트
categ = ['gr-qc', 'quant-ph', 'q-bio', 'nlin', 'physics', 'math-ph', 'q-fin', 'eess', 'hep-ex', 'cond-mat', 'math', 'econ', 'stat', 'nucl-ex', 'nucl-th', 'hep-lat', 'astro-ph', 'cs', 'hep-ph', 'hep-th']

# 결과를 저장할 딕셔너리 초기화, 각 카테고리에 대한 빈 리스트를 만들어준다.
result_dict = {category: [] for category in categ}

# 입력 리스트의 각 요소에 대해
for item in input_list:
    # 점(.)을 기준으로 분할하여 키와 값을 분리
    try:
        key, value = item.split('.', 1)
    except ValueError:
        # 점(.)이 없는 항목은 무시
        continue
    
    # 해당 키가 딕셔너리에 존재하는 경우에만 추가
    if key in result_dict:
        result_dict[key].append(item)

# 딕셔너리를 사용하여 각 카테고리에 대한 리스트를 추출하고 비어있지 않은 리스트만 필터링
non_empty_categories = {k: v for k, v in result_dict.items() if v}

non_empty_categories
```

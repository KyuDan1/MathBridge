```python
# GPT API 호출 함수 정의
def latex_to_spoken_english(latex_equation):
    messages = [
        {"role": "system", "content": 
        """
        You are a ChatGPT, a large language model trained by OpenAI, based on the GPT-3.5 architecture.
        Knowledge cutoff: 2022-01
        Current date: 2024-05-08
        Personality: v2
        
        You should only translate the equations without adding any extra explanations.
        """},
        {"role": "user", "content":
          f"""
          Translate the following LaTeX equation into spoken English without adding any extra explanations: {latex_equation}
        For example,

        LaTeX:$ x = \frac{{-b \pm \sqrt{{b^2 - 4ac}}}}{{2a}} $   ChatGPT:x equals negative b plus or minus the square root of b squared minus four a c all over two a

        LaTeX:$ a^2 + b^2 = c^2 $        ChatGPT:a squared plus b squared equals c squared

        If the equation is not a standard mathematical equation, output only 'None'
        """}
    ]
    response = openai.ChatCompletion.create(
        engine="gpt1",
        messages=messages,
        max_tokens=150
    )
    spoken_english = response.choices[0].message['content'].strip()
    return spoken_english

# spoken_English 열 추가
df_test['spoken_English'] = df_test['equation'].apply(latex_to_spoken_english)
df_test
```
